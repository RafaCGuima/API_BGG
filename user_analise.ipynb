# Puxando jogos do usuario do BGG

```python
import pandas as pd
```

```python
user = str(input("Digite o nome do usuário: "))
```

```python
df_basegame = pd.read_xml(f'https://api.geekdo.com/xmlapi2/collection?username={user}&own=1&excludesubtype=boardgameexpansion')
df_basegame
```

```python
df_expansions = pd.read_xml(f'https://api.geekdo.com/xmlapi2/collection?username={user}&own=1&subtype=boardgameexpansion')
df_expansions
```

## Numero de jogos base e de expansões na coleção

```python
print(f'Coleção do usuário {user} possui {df_basegame.shape[0]} jogos base e {df_expansions.shape[0]} expansões')
```

## Limpando os Dataframes

```python
df_basegame.drop(['objecttype', 'collid', 'image', 'thumbnail', 'status', 'originalname'], axis = 1, inplace=True)
df_expansions.drop(['objecttype', 'collid', 'image', 'thumbnail', 'status'], axis = 1, inplace=True)
```

## Encontrando o índice H

```python
games_by_plays = df_basegame.sort_values(by=('numplays'),ascending = False)
games_by_plays.index = range(games_by_plays.shape[0])
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3):
    display(games_by_plays)



```

```python
games_by_plays.info()
```

```python
indice = 0
for jogadas in games_by_plays['numplays']:
    if (indice  <= jogadas):
        indice += 1
        
print(f'Índice h da coleção do usuário {user} é igual a {indice}')
```

```python
proximo = games_by_plays.loc[[indice],['name','numplays']]
print(f'Próximo jogo a atingir o índice provavelmente será {proximo.name}')
```
